```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('stats_tourisme.csv')
df.head(5)
```

```python
df[df["Dépense internationale pour les voyages à l'étranger ($US2021)"].notna() & df["PIB / habitant"]]

depenses = df.loc[df["Dépense étrangère dans le pays pour le tourisme ($US2021)"].notna() & df["PIB / habitant"].notna(), ["Pays","Dépense étrangère dans le pays pour le tourisme ($US2021)","PIB / habitant","Population","PIB","Part du tourisme dans le PIB"]]
depenses=depenses[depenses["Pays"]!="Luxembourg"]
depenses=depenses[depenses["Pays"]!="Iceland"]
depenses=depenses[depenses["Pays"]!="Austria"]
depenses = depenses.sort_values(by="PIB")
depenses
```

```python
graph=depenses.plot.bar(x="Pays", y=["PIB","Dépense étrangère dans le pays pour le tourisme ($US2021)"], log=True, rot=90)
graph.set(ylim=(10**7,10**14), title="Comparaison logarithmique du des revenus dues au tourisme et du PIB", ylabel="$ US", xlabel="Pays")
plt.show()
```

```python
graph2= depenses.plot.bar(x="Pays", y="Part du tourisme dans le PIB")
```

```python
travail=df.loc[df["Nombre d'employés dans le tourisme pour 1000 habitants"].notna() & df["Nombre d'employés dans la restauration pour 1000 habitants"].notna(), ["Pays","Nombre d'employés dans la restauration pour 1000 habitants","Nombre d'employés dans le tourisme pour 1000 habitants"]]
travail["diff"]=travail["Nombre d'employés dans le tourisme pour 1000 habitants"]-travail["Nombre d'employés dans la restauration pour 1000 habitants"]
travail.plot.bar(x="Pays", y=["Nombre d'employés dans la restauration pour 1000 habitants","diff"], stacked=True, label=["dont employés du secteur de la restauration","employés du secteur du tourisme"], xlabel="Pays", title="nombre d'employés pour 100 habitants du pays")
```

Fonction permettant de tracer un histogrmme en fonction du pays.

```python
def histogramme(titre = None, X=None, Y=None, titrex=None, titrey=None, tri=None):
    L=[X,Y,tri]
    df1=df.loc[df[Y].notna() & df[X].notna(), L]
    if tri!=None:
        df1=df1.sort_values(by=tri)
    df1.plot.bar(x=X, y=Y, xlabel=titrex, ylabel=titrey, title=titre,  )
    plt.show()
    
```

```python
histogramme(X="Pays", Y="Nombre de départs à l'étranger par an pour 1000 habitants", titre="Nombre de départs à l'étranger par an pour 1000 habitants par Pays", titrex="Pays", tri="PIB / habitant")
```
